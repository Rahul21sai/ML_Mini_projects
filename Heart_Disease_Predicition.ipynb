```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#data link = https://drive.google.com/file/d/1CEql-OEexf9p02M5vCC1RDLXibHYE9Xz/view
data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
x = data.drop(columns='target', axis = 1)
y = data['target']
```

```python
print(x)
print(y)
```

```python
#train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the model
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predcition
t_predicit = model.predict(x_train)
```

```python
 train_data_accuracy = accuracy_score(t_predicit, y_train)
```

```python
print(train_data_accuracy*100)
```

```python
#test predicition
test_predicit = model.predict(x_test)
```

```python
test_data_accuracy = accuracy_score(test_predicit, y_test)
```

```python
print(test_data_accuracy)
```

```python
#predicitive system
input_data = (56,0,1,140,294,0,0,153,0,1.3,1,0,2)
idata_as_numpy = np.array(input_data)
input_reshaped = idata_as_numpy.reshape(1,-1)
predicition = model.predict(input_reshaped)
print(predicition)

if predicition == 0:
  print("the person doesnt have heart disease")
else :
  print("the person have heart disease")
```
